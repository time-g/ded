```python
import pandas as pd
import matplotlib.pyplot as plt
from ded import ded

plt.style.use('https://github.com/dhaitz/matplotlib-stylesheets/raw/master/pitayasmoothie-dark.mplstyle')
output_dir = 'img/'
```

# Estimating the Optimal Number of Clusters in aggregation dataset

```python
df_aggregation = pd.read_csv('data/Aggregation.txt', delimiter='\t', usecols=[0,1,2], header=None)
df_aggregation.head()
```

```python
df_aggregation.describe()
```

```python
df_aggregation.isna().sum()
```

## estimate optimal number of clusters

```python
ded(df_aggregation[[0,1]].to_numpy())
```

```python
color = {'green': '#5df579'}
plt.scatter(df_aggregation[0], df_aggregation[1], color=color['green'])
plt.savefig(f'{output_dir}aggregation-dataset.svg', format='svg')
```
